#### 테이블부터 생성

```python
import sqlite3

conn= sqlite3.connect('../../../code/QuQu-backend/db.sqlite3')
cur = conn.cursor()
```

```python
conn.execute('drop table IF EXISTS ququ_attributes')
conn.execute('drop table IF EXISTS ququ_brand')
conn.execute('drop table IF EXISTS ququ_category')
conn.execute('drop table IF EXISTS ququ_common_attributes')
conn.execute('drop table IF EXISTS ququ_hue')
conn.execute('drop table IF EXISTS ququ_tone')
conn.execute('drop table IF EXISTS ququ_attributes')
conn.execute('drop table IF EXISTS ququ_vogue')
conn.execute('drop table IF EXISTS ququ_cat_attr')
conn.execute('drop table IF EXISTS ququ_cat_attr_encoding')
conn.commit()
```

```python
conn.execute('''CREATE TABLE "ququ_attributes" (
"index" INTEGER primary key AUTOINCREMENT,
"id" INTEGER,
  "name" TEXT
)''')

conn.execute('''CREATE TABLE "ququ_brand" (
"index" INTEGER primary key AUTOINCREMENT,
  "brands" TEXT,
  "year" INTEGER,
  "season" TEXT,
    "image" TEXT
)''')

conn.execute('''CREATE TABLE "ququ_category" (
"index" INTEGER primary key AUTOINCREMENT,"id" INTEGER,
  "name" TEXT
)''')

conn.execute('''CREATE TABLE "ququ_common_attributes" (
"index" INTEGER primary key AUTOINCREMENT,"year" INTEGER,
  "season" TEXT,
  "attributes" TEXT
)''')

conn.execute('''CREATE TABLE "ququ_hue" (
"index" INTEGER primary key AUTOINCREMENT,"hue_cluster" INTEGER,
  "top_index" TEXT,
  "year" TEXT,
  "season" TEXT
)''')


conn.execute('''CREATE TABLE "ququ_tone" (
"index" INTEGER primary key AUTOINCREMENT,"tone_cluster" INTEGER,
  "top_index" TEXT,
    "year" TEXT,
  "season" TEXT
)''')

conn.execute('''CREATE TABLE "ququ_vogue" (
"index" INTEGER primary key AUTOINCREMENT,"image" TEXT,
  "category" INTEGER,
  "attributes" TEXT,
  "year" INTEGER,
  "season" TEXT,
  "brand" TEXT
)''')

conn.execute('''CREATE TABLE "ququ_cat_attr" (
"index" INTEGER primary key AUTOINCREMENT,
  "category" INTEGER,
  "attribute" TEXT,
  "year" INTEGER,
  "season" TEXT
)''')
conn.commit()
```
